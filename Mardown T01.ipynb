## Introducción a transferencia de calor por conducción

Se sobre los siguientes 3 tipos de transferencia de calor presentados en una lista ordenada del que me parece más simple al más complejo.

2. Convección
3. Radiación

Siendo la conducción el que me parece mas simple de todos y del que daré una pequeña introducción.

Este tipo de transferencia se da por contacto directo de dos cuerpos sin interferencia. Para obtener una conductividad mayor se usan materiales como metales, grafito o diamante, cuya estructura o presencia de electrones libres permite conducción óptima; o se previene con materiales aislantes como lo son la madera, los plásticos o el vidrio que tienen una estructura que impide el intercambio de energía y falta de esos electrones libres.

Para calcular el balance de energía que se da por esta forma de transferencia a través de una pared se usa la ecuación $Q=-k*A*dt/dx$, donde $Q$ es el calor, $k$ el coeficiente propio del material, $A$ el Área de pared que se estudia, $dT$ el cambio de temperatura observado (siempre viaja el calor del punto más caliente al más frío) y $dx$ el espesor de la pared en cuestión.

Saber de que material esta hecho el límite de la pared del sistema y las condiciones tanto dentro como fuera de el te permite tener una idea de por ejemplo en el caso de un acondicionador de aire o un reactor cuanta energía se necesitará para mantener la temperatura constante y óptima.
